```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import datetime as dt
import scipy.stats as stats
import math
```

```python
rates_unclean = pd.read_csv('rates.csv', parse_dates=['Dates'])
```

```python
rates_unclean0 = rates_unclean.sort_values(by=['Dates'])
rates_unclean0 = rates_unclean.set_index('Dates')
```

```python
rates_unclean0.head()
```

```python
rates_unclean1 = rates_unclean0.drop(['0.25', '0.75'], axis=1)
rates_unclean1.head()
```

```python
np.arange(16, 19, 1)
```

```python
seq_1 = np.arange(1.5, 18.5, 1)
seq_2 = np.arange(4, 5, 1)
seq_3 = np.arange(6, 7, 1)
seq_4 = np.arange(8, 10, 1)
seq_5 = np.arange(11, 15, 1)
seq_6 = np.arange(16, 19, 1)
```

```python
new_cols_float = sorted(np.concatenate([seq_1, seq_2, seq_3, seq_4, seq_5, seq_6], axis=0))
```

```python
new_cols_str = ['{:.2f}'.format(x) for x in new_cols_float]
new_cols_str[:5]
```

```python
new_vals = [np.nan]*len(new_cols_str)

rates_unclean2 = rates_unclean1.reindex(columns=rates_unclean1.columns.tolist() + new_cols_str)   # add empty cols
rates_unclean2[new_cols_str] = new_vals  # multi-column assignment works for existing cols
```

```python
rates_unclean2.head()
```

```python
sorted(rates_unclean2.columns)[:10]
```

```python
rates_unclean3 = rates_unclean2.reindex(['{:.2f}'.format(x) for x in sorted(rates_unclean2.columns.astype('float'))], 
                                        axis=1)
rates_unclean3.head()
```

```python
rates_clean = rates_unclean3.interpolate(method = 'linear',
                                         axis = 1) # берем среднее между соседними ставками для пропусков
rates_clean.head()
```

```python
len(rates_clean.columns) == 18*2
```

```python
rates_clean.to_csv('rates_clean.csv')
```
